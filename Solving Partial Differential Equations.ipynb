```python
import math 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from jupyterthemes import jtplot
jtplot.style(theme='monokai', context='notebook', ticks=True, grid=True)
```

## Elliptical PDE's

```python
n = 15
V = np.zeros((n,n))
V[0,:] = 100   
tolerance = .001
V_old = np.sum(V**2)

for i in range(1,n-1):
    for j in range(1,n-1):
        V[i,j] = (V[i+1,j] + V[i-1,j] + V[i,j+1] + V[i,j-1])/4
V_new = np.sum(V**2)

while (abs(V_new - V_old)>=tolerance):
    V_old = V_new
    for i in range(1,n-1):
        for j in range(1,n-1):
            V[i,j] = (V[i+1,j] + V[i-1,j] + V[i,j+1] + V[i,j-1])/4
    V_new = V[int(n/2),int(n/2)]
```

```python
plt.figure(figsize=(10,10))
plt.imshow(V)
plt.show()
```

```python
x, y = np.meshgrid(np.linspace(0, n - 1, n), np.linspace(n-1, 0, n))
fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot(111, projection = '3d')
ax.plot_surface(x, y, V)
```

## Parabolic PDEs

```python
x_total = 40
t_total = 1000
T = np.zeros((x_total, t_total))
T[:,0] = 100  
eta = .35

for j in range(0,t_total-1):
    for i in range(1,x_total-1):
        T[i,j+1] = T[i,j] + eta * (T[i+1,j] + T[i-1,j] - 2*T[i,j])

plt.figure(figsize=(10,10))
plt.imshow(T)
plt.show()
```

```python
x, t = np.meshgrid(np.linspace(0, x_total - 1, x_total), np.linspace(0, t_total - 1, t_total))
fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot(111, projection = '3d')
ax.plot_surface(t, x, T.T)
ax.set_ylabel("Points along the length of wire")
ax.set_xlabel("Time Axis")
ax.set_zlabel("Temperature in degree Celcius")
print(len(x))
```
