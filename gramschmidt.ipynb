```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.integrate as sint
```

```python
n=5
m=100
x=np.linspace(-1,1,m)
v=np.zeros((n,m))
for i in range(0,n):
    v[i]= x**i
```

```python
def inner_prod(u,v):
    return sint.simpson(u*v,dx=(x[-1]-x[0])/len(x))
```

```python
u=np.zeros((n,m))
u[0]=v[0]/(inner_prod(v[0],v[0])**0.5)
for i in range(1,n):
    temp=v[i]
    for j in range(0,i):
        temp -= (inner_prod(v[i],u[j])*u[j])
    u[i]=temp/inner_prod(temp,temp)**0.5
```

```python
#plt.style.use('dark_background')

for i in range(n):
    plt.plot(x,u[i])
```

```python
exponent=np.exp(x)
print(exponent)
```

```python
ap=0
for i in range(0,n):
    ap+= inner_prod(exponent,u[i])*u[i]
    
```

```python
plt.plot(x,ap)

plt.plot(x,exponent)
```
