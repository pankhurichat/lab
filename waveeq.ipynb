```python
import matplotlib.pyplot as plt
import numpy as np
#%matplotlib tk
```

```python
x=100 
y=1000
pos=np.zeros((x,y))
```

```python
#Initial Conditions
time=np.linspace(0,1,x)
for i in range(len(time)):
    if time[i]<=.8:
        pos[i,0]=.125*time[i]
    else:
        pos[i,0]=.5-.5*time[i]
```

```python
#n=1
for i in range(1,x-1):
    pos[i,1]=pos[i,0]+(1/2)*(pos[i+1,0]+pos[i-1,0]-2*pos[i,0])

for j in range(1,y-1):
    for i in range(1,x-1):
        pos[i,j+1]=2*pos[i,j]+ -pos[i,j-1]+(pos[i+1,j]+pos[i-1,j]-2*pos[i,j]) 
```

```python
for i in range(y-1):
    plt.plot(np.linspace(0,1,100),pos[:,i])
```
